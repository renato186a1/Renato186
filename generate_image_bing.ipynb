```python
# prompt: generate images bing and display screen

!pip install bing-image-downloader

from bing_image_downloader import downloader
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# Replace 'your_search_term' with the actual search term
search_term = 'imagens ia hulk'

# Download images (adjust limit as needed)
downloader.download(search_term, limit=10,  output_dir='downloaded_images', adult_filter_off=True, force_replace=False, timeout=60)

# Display downloaded images
image_dir = 'downloaded_images/' + search_term

import os
for filename in os.listdir(image_dir):
  if filename.endswith(('.jpg', '.jpeg', '.png')):  # Adjust extensions if necessary
    image_path = os.path.join(image_dir, filename)
    try:
        img = mpimg.imread(image_path)
        plt.figure(figsize=(8, 6))  # Adjust figure size as needed
        plt.imshow(img)
        plt.axis('off')  # Hide axes
        plt.show()
    except Exception as e:
        print(f"Error displaying image {filename}: {e}")
```
